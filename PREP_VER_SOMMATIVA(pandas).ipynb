1. l'ammontare del reddito da lavoro dipendente (colonna L del dataset) per ogni regione e relativo grafico a torta

```python
ammonto_regioni = redditi.groupby("Regione")[['Reddito da lavoro dipendente e assimilati - Ammontare in euro']].sum()
ammonto_regioni
```

```python
#plot = ammonto_regioni.plot.pie(y="Reddito da lavoro dipendente e assimilati - Ammontare in euro",figsize=(20,20))

labels = list(ammonto_regioni.index)
sizes = list(ammonto_regioni["Reddito da lavoro dipendente e assimilati - Ammontare in euro"])
fig1, ax1 = plt.subplots(figsize=(10, 10))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',textprops={'fontsize': 14}, startangle= 90)
plt.title("l'ammontare del reddito da lavoro dipendente ")
plt.show()
```

2. l'ammontare del reddito da lavoro dipendente per ogni provincia e relativo grafico a barre verticali

```python
reddito_prov = pd.merge(redditi, province, how = "inner", left_on=["Sigla Provincia"], right_on=["SIGLA"])
reddito_prov
```

```python
ammonto_province = reddito_prov.groupby("DEN_UTS").sum()
ammonto_province[["Reddito da lavoro dipendente e assimilati - Ammontare in euro"]]
```

```python
#ammonto_province.plot.bar(y="Reddito da lavoro dipendente e assimilati - Ammontare in euro",figsize=(25,10))

x = list(ammonto_province.index)
y = list(ammonto_province["Reddito da lavoro dipendente e assimilati - Ammontare in euro"])

fig,ax = plt.subplots()
fig.set_figwidth(30)
fig.set_figheight(10)
ax.set_title("AMMONTARE REDDITO LAVORO DIPENDENTE PER OGNI PROVINCIA", fontsize=20)
ax.set_xlabel("PROVINCIA", fontsize=15)
ax.set_ylabel("AMMONTARE REDDITO LAVORO DIPENDENTE", fontsize=15)
fig.autofmt_xdate(rotation=90)
ax.bar(x,y)
```
